```python
# Anaísa Forti da Fonseca
# 11811ECP012

import importlib
if importlib.util.find_spec("control") is None:
    !pip install control

import numpy as np
import matplotlib.pyplot as plt
import control as ctl

def graph_size(n):
    """Função auxiliar para definir tamanho dos gráficos"""
    return (n*(1+5**0.5)/2, n)
```

```python
# X(s) = (s^2 + s + 4) / (s^4 + 5s^2 + 4)

t = np.linspace(0., 1.5, 100)
sys = ctl.tf([1.,1.,4.],[1.,0.,5.,0.,4.])
T, yout = ctl.impulse_response(sys, T=t)

fig = plt.figure(figsize=graph_size(7))
plt.plot(T, yout, label="simulada")
plt.grid('on')
plt.xlabel('Tempo $t$')
plt.legend()

print("Zeros: ", sys.zero())
print("Polos: ", sys.pole())
```
