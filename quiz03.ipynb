```python
# Anaísa Forti da Fonseca
# 11811ECP012

import numpy as np
import scipy.signal as sgn
import scipy as sc
import matplotlib.pyplot as plt
import control as ctl
from IPython.display import display
from ipywidgets import interact, fixed

def graph_size(n):
    """Função auxiliar para definir tamanho dos gráficos"""
    return (n*(1+5**0.5)/2, n)
```

```python
def show_poles():
    G = ctl.tf([1.], [1., 1/2., 1., 0])
    poles = np.poly1d(G.den[0][0]).r
    
    def ROC(G, inf=100):
        poles = np.poly1d(G.den[0][0]).r
        minpole, maxpole = np.real(np.max(poles)), inf
        plt.fill_between([minpole, maxpole], [-inf, -inf], [inf, inf], color='g', alpha=0.5)
    
    fig = plt.figure(figsize=graph_size(7))
    for p in poles:
        plt.scatter(np.real(p), np.imag(p), 144, marker='x', color='r')
    ROC(G)
    plt.grid('on')
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.xlim([-1,1])
    plt.ylim([-2,2])
show_poles()
# reposta: gráfico marginalmente estável
```
